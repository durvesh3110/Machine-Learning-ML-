1. **What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.**

**Ans:**

1. Artificial Intelligence (AI)

    - A broad field of computer science focused on creating machines that can perform tasks that normally require human intelligence.

    - Examples of tasks: Decision-making, language understanding, problem-solving, and visual perception.

    - Applications: Chatbots, self-driving cars, recommendation systems.

2. Machine Learning (ML)

    - A subset of AI where machines learn patterns from data and improve performance without being explicitly programmed for every task.

    - How it works: Algorithms analyze data and make predictions or decisions.

    - Examples: Spam email detection, credit card fraud detection, movie recommendations.

3. Deep Learning (DL)

    - A subset of ML that uses neural networks with many layers to learn complex patterns.

    - It is inspired by the structure of the human brain.

    - Examples: Image recognition, speech recognition, autonomous driving systems.

4. Data Science

    - An interdisciplinary field that uses statistics, programming, and data analysis to extract insights from data.

    - Tools used: Python, R, SQL, machine learning models, data visualization.

    - Examples: Business analytics, customer behavior analysis, forecasting.

2. **Explain overfitting and underfitting in ML. How can you detect and prevent them?**

**Ans:**

1. Overfitting

    - Overfitting occurs when a machine learning model learns the training data too well, including noise and random fluctuations. As a result, it performs very well on training data but poorly on new or unseen data.
    - Detection:
        - Large gap between training accuracy and validation accuracy
        - Training loss is very low while validation loss is high
        - Seen during model evaluation using validation or test datasets

    - Prevention:
        - Use more training data
        - Apply regularization (L1, L2)
        - Use dropout (in neural networks)
        - Simplify the model (reduce parameters/features)
        - Use cross-validation
        - Early stopping during training

2. Underfitting

    - Underfitting occurs when a model is too simple to capture the underlying patterns in the data. It fails to perform well on both training and new data.

    - Detection:
      - Both training accuracy and validation accuracy are low
      - Loss remains high even after training

    - Prevention:
      - Use a more complex model
      - Add more features
      - Train the model for more iterations
      - Reduce regularization strength
      - Use better feature engineering

3. **How would you handle missing values in a dataset? Explain at least three methods with examples.**

**Ans:**

1. Removing Missing Values (Deletion Method)

    In this method, rows or columns containing missing values are removed from the dataset.

    - Example:

      | Name | Age | Salary |
      | ---- | --- | ------ |
      | A    | 25  | 30000  |
      | B    | NA  | 35000  |
      | C    | 28  | 40000  |

      After removing the row with missing value:

      | Name | Age | Salary |
      | ---- | --- | ------ |
      | A    | 25  | 30000  |
      | C    | 28  | 40000  |


2. Mean / Median / Mode Imputation

    Missing values are replaced with a statistical value from the column.

    | Age |
    | --- |
    | 25  |
    | NA  |
    | 30  |
    | 35  |

    Mean = (25 + 30 + 35) / 3 = 30

    After replacement:

    | Age |
    | --- |
    | 25  |
    | 30  |
    | 30  |
    | 35  |


3. Prediction-Based Imputation

    Missing values are predicted using machine learning models or relationships between variables.

    - Example:

      | Age | Experience | Salary |
      | --- | ---------- | ------ |
      | 25  | 2          | 30000  |
      | 30  | 5          | 45000  |
      | 35  | 8          | NA     |

      Using regression based on Age and Experience, the model predicts the missing Salary value.

      Predicted Salary ≈ 50000


4. **What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).**

**Ans:**

  - Imbalanced Dataset

    An imbalanced dataset occurs when the number of observations in one class is much higher than the other class(es) in a classification problem. This imbalance can cause machine learning models to become biased toward the majority class, reducing their ability to correctly predict the minority class.

  - Two techniques to handle Imbalanced Dataset:

1.   Oversampling

  - Theoretical Explanation
    - Oversampling increases the number of samples in the minority class so that both classes become more balanced. This helps the model learn patterns from the minority class better.

    - A popular method is SMOTE, which creates synthetic samples instead of simply duplicating existing ones.

  - Practical Example

    | Class   | Samples |
    | ------- | ------- |
    | Class A | 900     |
    | Class B | 100     |

    After oversampling:

    | Class   | Samples |
    | ------- | ------- |
    | Class A | 900     |
    | Class B | 900     |



2.   Undersampling

  - Theoretical Explanation

    - Undersampling reduces the number of samples in the majority class to balance the dataset. This is done by randomly removing some majority class samples.

  - Practical Example

    | Class   | Samples |
    | ------- | ------- |
    | Class A | 900     |
    | Class B | 100     |

    After undersampling:

    | Class   | Samples |
    | ------- | ------- |
    | Class A | 100     |
    | Class B | 100     |


5. **Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.**

**Ans:**

  - The feature scaling is important in ML because:
    - Improves model performance
    - Prevents bias toward large values
    - Required for distance-based algorithms
    - Faster training

  - Comparing Min-Max and Standardization:

    | Aspect                  | Min–Max Scaling             | Standardization      |
    | ----------------------- | --------------------------- | -------------------- |
    | Range                   | 0 to 1 (or fixed range)     | Mean = 0, Std = 1    |
    | Formula                 | (X − min) / (max − min)     | (X − mean) / std     |
    | Sensitivity to Outliers | High                        | Lower                |
    | Best Used For           | Neural networks, image data | Regression, SVM, PCA |


6. **Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?**

**Ans:**

| Aspect           | Label Encoding              | One-Hot Encoding        |
| ---------------- | --------------------------- | ----------------------- |
| Output           | Single column with integers | Multiple binary columns |
| Order assumption | Yes                         | No ordering             |
| Memory usage     | Low                         | Higher                  |
| Suitable for     | Ordinal data                | Nominal data            |


  - When to use which?

    - Label Encoding:
      - When the categorical data is ordinal.
      - Example: Low, Medium, High

    - One-Hot Encoding:
      - When the categorical data is nominal.
      - Example: Colors, Cities, Countries


7. **Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?**

**Ans:**

```python
import pandas as pd

# Load dataset
df = pd.read_csv('googleplaystore.csv')

# Convert rating to numeric and remove missing values
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df_clean = df.dropna(subset=['Rating', 'Category'])

# Calculate average rating per category
avg_ratings = df_clean.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Top and bottom categories
top5 = avg_ratings.head(5)
bottom5 = avg_ratings.tail(5)

print("Top 5 Categories by Average Rating:\n")
print(top5)

print("\nBottom 5 Categories by Average Rating:\n")
print(bottom5)
```

9. Flight Price Prediction Dataset

a) **How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.**

**Ans:**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('flight_price.csv')

# Filter for Delhi-Mumbai route
delhi_mumbai = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Calculate average price by days left
price_by_days = delhi_mumbai.groupby('days_left')['price'].mean().reset_index()
price_by_days = price_by_days.sort_values('days_left')

# Plot
plt.figure(figsize=(12, 6))
plt.plot(price_by_days['days_left'], price_by_days['price'], marker='o', linewidth=2)
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (₹)')
plt.title('Flight Price Trend by Days Left (Delhi-Mumbai)')
plt.grid(True, alpha=0.3)
plt.axvspan(20, 50, alpha=0.2, color='green', label='Optimal Window (20-50 days)')
plt.axvspan(1, 9, alpha=0.2, color='red', label='Surge Zone (1-9 days)')
plt.legend()
plt.tight_layout()
plt.show()

# Calculate key statistics
avg_30_50 = price_by_days[(price_by_days['days_left'] >= 30) & (price_by_days['days_left'] <= 50)]['price'].mean()
avg_1_9 = price_by_days[(price_by_days['days_left'] >= 1) & (price_by_days['days_left'] <= 9)]['price'].mean()

print(f"Average price (30-50 days before): ₹{avg_30_50:,.2f}")
print(f"Average price (1-9 days before): ₹{avg_1_9:,.2f}")
print(f"Price increase: {((avg_1_9/avg_30_50)-1)*100:.1f}%")
print(f"\nBest booking window: 20-50 days before departure")
```

b) **Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?**

**Ans:**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('flight_price.csv')

# Filter for Delhi-Mumbai route
delhi_mumbai = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Calculate average price by airline
airline_prices = delhi_mumbai.groupby('airline')['price'].agg(['mean', 'count']).round(2)
airline_prices = airline_prices.sort_values('mean')

print("Average Price by Airline (Delhi-Mumbai):")
print(airline_prices)
print()

# Plot
plt.figure(figsize=(10, 6))
colors = ['green' if i < 3 else 'red' if airline == 'Vistara' else 'orange'
          for i, airline in enumerate(airline_prices.index)]
airline_prices['mean'].plot(kind='bar', color=colors)
plt.ylabel('Average Price (₹)')
plt.title('Average Flight Prices by Airline (Delhi-Mumbai)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Simple analysis
budget_avg = airline_prices.loc[['AirAsia', 'GO_FIRST', 'SpiceJet'], 'mean'].mean()
vistara_avg = airline_prices.loc['Vistara', 'mean']

print(f"Budget airlines average: ₹{budget_avg:,.2f}")
print(f"Vistara average: ₹{vistara_avg:,.2f}")
print(f"Vistara is {(vistara_avg/budget_avg-1)*100:.1f}% more expensive")
print("\nCheapest: AirAsia, GO_FIRST, SpiceJet (budget airlines)")
print("Most expensive: Vistara (premium full-service airline)")
```
