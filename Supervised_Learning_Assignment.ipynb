1. **What is Simple Linear Regression (SLR)? Explain its purpose.**

**Ans:**

  - Simple Linear Regression finds the best-fitting straight line that explains how a dependent variable changes when the independent variable changes.
  - Its Purpose:

    - Predict values

      Estimate the value of a dependent variable based on a known independent variable.

      Example: Predicting house price from house size.

    - Understand relationships

      Determine whether and how strongly two variables are related.

      Example: Relationship between study hours and exam scores.

    - Trend analysis

      Identify patterns or trends in data over time or across variables.

    - Decision making

      Helps businesses and researchers make decisions using predictive insights.

2. **What are the key assumptions of Simple Linear Regression?**

**Ans:**



1.   Linearity

       There must be a linear relationship between the independent variable (X) and the dependent variable (Y).

        - The change in Y should be proportional to the change in X.
        - A scatter plot of X and Y should show a roughly straight-line pattern.

      Example: As study hours increase, exam scores increase linearly.

2.   Independence of Errors

       The residuals (errors) must be independent of each other.

        - The error for one observation should not influence another.
        - This is important in time-series data, where consecutive values might be correlated.

3.   Homoscedasticity (Constant Variance)

       The variance of the errors should be constant for all values of X.

        - Residuals should have equal spread across the regression line.
        - If the spread increases or decreases with X, it is called heteroscedasticity, which violates the assumption.

4.   Normality of Errors

       The residuals should follow a normal distribution.

        - This assumption is important for hypothesis testing and confidence intervals.
        - It can be checked using histograms or normal probability plots.

5.   No Measurement Error in the Independent Variable

       The independent variable (X) is assumed to be measured without error or with negligible error.

        - If X contains large measurement errors, the regression estimates can become biased.



3. **Write the mathematical equation for a simple linear regression model and explain each term.**

**Ans:**

Mathematical equation for simple linear equation:

Y = a + bX + e

Explanation of Each Term

1. Y (Dependent Variable)
    - The variable whose value we want to predict or explain.
    - It depends on the independent variable.
    - Example: exam score, sales, house price.

2. a (Intercept (Constant Term))
    - The value of Y when X = 0.
    - It represents the point where the regression line cuts the Y-axis.

3. b (Slope (Regression Coefficient))
    - Shows the rate of change in Y for a one-unit increase in X.

      If b is positive, Y increases as X increases.

      If b is negative, Y decreases as X increases.

4. X (Independent Variable)
    - The variable used to predict or explain Y.
    - It is sometimes called the predictor or explanatory variable.
    - Example: hours studied, advertising expenditure, size of a house.

5. e (Error Term (Residual))
    - Represents the difference between the observed value and the predicted value.
    - It captures other factors affecting Y that are not included in the model.

5. **What is the method of least squares in linear regression?**

**Ans:**

  - The method of least squares finds the regression line by minimizing the sum of the squares of the differences between the observed values and the predicted values.

  - Principle of Least Squares

    For each data point, the residual (error) is:

    Error = Observed value − Predicted value

    The method minimizes:

    ∑(𝑌−𝑌^)^2


  - This means it minimizes the sum of squared residuals, ensuring the regression line is the closest possible fit to all data points.

7. **Name and briefly describe three common evaluation metrics for regression models.**

**Ans:**

  - Three common evaluation metrics for regression models:


1.   Mean Absolute Error (MAE)

  - Mean Absolute Error measures the average absolute difference between the actual values and the predicted values.

    MAE = 1/n 	​∑ ∣ Y − Y^ ∣

    - It calculates the average magnitude of errors without considering their direction.
    - Lower MAE indicates better model performance.
    - It is easy to interpret because it uses the same units as the data.

2.   Mean Squared Error (MSE)

  - Mean Squared Error measures the average of the squared differences between actual and predicted values.

    MSE = 1/n ​∑(Y−Y^)2

    - Squaring the errors penalizes larger errors more heavily.
    - Often used in optimization during model training.
    - Lower MSE means the model has better predictive accuracy.

3.   Root Mean Squared Error (RMSE)

  - Root Mean Squared Error is the square root of the Mean Squared Error.

    RMSE = square root(1/n	​∑(Y−Y^)2)

    - It measures the standard deviation of prediction errors.
    - RMSE is in the same unit as the dependent variable, making it easier to interpret.
    - Lower RMSE indicates better model performance.

8. **What is the purpose of the R-squared metric in regression analysis?**

**Ans:**

  - The R-squared metric, also called the coefficient of determination, is used in regression analysis to measure how well the regression model explains the variation in the dependent variable.

  - Purpose of R-squared:
    - The main purpose of R² is to evaluate the goodness of fit of a regression model.
    - It shows what proportion of the total variation in the dependent variable (Y) is explained by the independent variable(s) (X) in the model.

    The formula for R-squared is:

    R^2 = 1 - \frac{SS_{res}}{SS_{tot}}

    Where:

    SS₍res₎ = Sum of squares of residuals (unexplained variation)

    SS₍tot₎ = Total sum of squares (total variation in the data)

  - Interpretation

    R² = 1 → Perfect fit (model explains 100% of the variation).

    R² = 0 → Model explains none of the variation.

    0 < R² < 1 → Indicates the proportion of variance explained by the model.

  - Example:

    If R² = 0.80, it means 80% of the variation in the dependent variable is explained by the independent variable(s), while 20% remains unexplained.

9. **Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.**

**Ans:**

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Example data (independent variable X and dependent variable y)
X = np.array([[1], [2], [3], [4], [5]])   # Feature (must be 2D)
y = np.array([2, 4, 5, 4, 5])             # Target values

# Create the regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Get slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print results
print("Slope (Coefficient):", slope)
print("Intercept:", intercept)
```

10. **How do you interpret the coefficients in a simple linear regression model?**

**Ans:**

  - In a Simple Linear Regression model, the coefficients explain how the independent variable affects the dependent variable.
  - Interpretation of the Coefficients:
    - Intercept (a)
      - The intercept represents the predicted value of Y when X = 0.
      - Example:

        Score = 40 + 5 × StudyHours

        The intercept 40 means that when StudyHours = 0, the predicted exam score is 40 marks.

    - Slope (b)
      - The slope indicates the change in the dependent variable (Y) for a one-unit increase in the independent variable (X).

        If b > 0 → Y increases as X increases (positive relationship).

        If b < 0 → Y decreases as X increases (negative relationship).

        If b = 0 → No relationship between X and Y.

      - Example:

        Score = 40 + 5 × StudyHours
        
        The slope 5 means that for every additional hour studied, the exam score increases by 5 marks on average.
